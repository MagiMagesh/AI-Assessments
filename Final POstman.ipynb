```python
import pandas as pd
import numpy as np
features = pd.read_csv('features.csv')
labels = pd.read_csv('labels.csv')
```

```python
features = features.dropna()
```

```python
features.isnull().sum()
```

```python
labels.dropna()
```

```python
labels.isnull().sum()
```

```python
val = labels['genre'].value_counts().sort_values().index
val
```

```python
codi = {k:i for i,k in enumerate(val,0)}
codi
```

```python
labels['mapping'] = labels['genre'].map(codi)
```

```python
labels = labels.drop(['genre'],axis=1)
```

```python
features.isnull().sum()
```

```python
features.dropna().head()
```

```python
features.isnull().sum()
```

```python
features = features.drop(['title','tags'],axis=1)
features.head()
```

```python
df = pd.merge(features,labels,how='outer')
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,:-1]
X.head(2)
```

```python
X = X.drop(['trackID'],axis=1)
```

```python
X.head(2)
```

```python
y = df.iloc[:,-1]
y.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor,ExtraTreesClassifier
selection = ExtraTreesRegressor()
selection.fit(X, y)
```

```python
#plot graph of feature importances for better visualization
import matplotlib.pyplot as plt
plt.figure(figsize = (12,8))
features_importances = pd.Series(selection.feature_importances_, index=X.columns)
features_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
from sklearn.ensemble import ExtraTreesRegressor,ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
ranked_feature = pd.Series(model.feature_importances_,index =  X.columns)
a = list(ranked_feature.nlargest(15).index)
a
```

```python
import seaborn as sns
corre = df[a].corr()
corre
```

```python
sns.pairplot(corre)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize =(10,10))
sns.heatmap(corre)
```

```python
X = X[a]
X.head(2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=50)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
print(X_test.shape,y_pred.shape,y_test.shape)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_test,y_pred))
```

```python
import pickle
import joblib
file = open('model.pkl','wb')

pickle.dump(classifier,file)
file.close()
```
